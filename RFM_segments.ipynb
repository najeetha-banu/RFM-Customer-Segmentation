```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Preview the dates
print(df['InvoiceDate'].head())
```

```python
# Create a TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Preview the new column
df[['Quantity', 'UnitPrice', 'TotalPrice']].head()
```

```python
import datetime as dt

# Set snapshot date to one day after the last invoice date
snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)

# Group by CustomerID to calculate RFM metrics
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'TotalPrice': 'sum'  # Monetary
})

# Rename columns
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Preview RFM table
rfm = rfm.reset_index()
print(rfm.head())
```

```python
# Import libraries
import numpy as np

# Define RFM score function
def rfm_score(x, quantiles):
    if x <= quantiles[0.25]:
        return 1
    elif x <= quantiles[0.50]:
        return 2
    elif x <= quantiles[0.75]:
        return 3
    elif x <= quantiles[0.95]:
        return 4
    else:
        return 5

# Calculate quantiles
quantiles = rfm.quantile(q=[0.25, 0.50, 0.75, 0.95])

# Apply scores
rfm['R_Score'] = rfm['Recency'].apply(rfm_score, quantiles=quantiles['Recency'])
rfm['F_Score'] = rfm['Frequency'].apply(rfm_score, quantiles=quantiles['Frequency'])
rfm['M_Score'] = rfm['Monetary'].apply(rfm_score, quantiles=quantiles['Monetary'])

# Combine RFM scores into a single string (e.g., "155")
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

print(rfm[['CustomerID', 'RFM_Segment']].head())
```

```python
# Define customer segments based on RFM scores
def rfm_segment(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 4 and row['M_Score'] >= 4:
        return 'Champions'
    elif row['R_Score'] >= 4 and row['F_Score'] >= 3:
        return 'Loyal Customers'
    elif row['R_Score'] >= 3:
        return 'Potential Loyalists'
    elif row['R_Score'] == 1 and row['F_Score'] == 1:
        return 'Lost Customers'
    else:
        return 'Others'

rfm['Customer_Segment'] = rfm.apply(rfm_segment, axis=1)

# Preview segments
print(rfm[['CustomerID', 'RFM_Segment', 'Customer_Segment']].head())
```
